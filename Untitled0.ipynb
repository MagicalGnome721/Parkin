```python
x_rows=['MDVP:Fo(Hz)','MDVP:Fhi(Hz)','MDVP:Flo(Hz)',
        'MDVP:Jitter(%)','MDVP:Jitter(Abs)']
```

```python
y_rows=['status']
```

```python
# Doc du lieu train
train_data = pd.read_csv('/content/drive/MyDrive/DoAn/Data/Data_Parkinsons_TRAIN.csv')
train_x = train_data[x_rows]
train_y = train_data[y_rows]
print("train_x:\n", train_x)
print("train_y:\n", train_y)
```

```python
# Su dung sklearn Gaussian Naive Bayes 
from sklearn.naive_bayes import GaussianNB 

gnb = GaussianNB() 
gnb.fit(train_x, train_y) 
```

```python
# Du doan train data
predict_train = gnb.predict(train_x)
print('Prediction on train data:', predict_train) 
```

```python
# Time diem chinh xac trong train data
from sklearn.metrics import accuracy_score
accuracy_train = accuracy_score(train_y, predict_train)
print('Accuracy score on train data:', accuracy_train)
```

```python
# Doc du lieu test data
test_data = pd.read_csv('/content/drive/MyDrive/DoAn/Data/Data_Parkinsons_TEST.csv')
test_x = test_data[x_rows]
test_y = test_data[y_rows]
```

```python
# Du doan test data
predict_test = gnb.predict(test_x)
print('Prediction on test data:', predict_test) 
```

```python
# Ti le diem chinh xac trong test data
accuracy_test = accuracy_score(test_y, predict_test)
print('Accuracy score on test data:', accuracy_train)
```
